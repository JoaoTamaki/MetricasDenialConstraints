```python
def logMetabase_to_dataframe(DCtxt, n_DCs):
    print('Generating dataframe...')

    DCs = pd.DataFrame(columns=['id', 'Tamanho', 'col1', 'op1', 'col2', 'op2', 'col3', 'op3', 'col4', 'op4'])
    
    DClist = []
    offset = 32
    aux_parse, aux_DC = parseDCFinderLog(DCtxt, n_DCs)
    for i in range(n_DCs):
        DClist.append( ( DCtxt[i].split('{"type":"DenialConstraint","predicates":[')[1] ).split(',') )
        tam = int(len(DClist[i])/8)

        DCcol = []
        DCop = []
        col_i = 13
        op_i = 19

        for j in range(1,tam+1):
            if (tam >= j):
                DCcol.append( (DCtxt[i].split('{"type":"DenialConstraint","predicates":[')[1] ).split('"')[col_i] )
                DCop.append( (DCtxt[i].split('{"type":"DenialConstraint","predicates":[')[1] ).split('"')[op_i] )
                col_i += offset
                op_i += offset
        
        if(tam == 1):
            new_rows = pd.DataFrame( {'id':[i], 'Tamanho':[tam], 'col1':[DCcol[0]], 'op1':[DCop[0]], 'col2':[0], 'op2':[0], 'col3':[0], 'op3':[0], 'col4':[0], 'op4':[0]} )
            DCs = pd.concat([DCs, new_rows])
        elif(tam == 2):
            new_rows = pd.DataFrame( {'id':[i], 'Tamanho':[tam], 'col1':[DCcol[0]], 'op1':[DCop[0]], 'col2':[DCcol[1]], 'op2':[DCop[1]], 'col3':[0], 'op3':[0], 'col4':[0], 'op4':[0]} )
            DCs = pd.concat([DCs, new_rows])
        elif(tam == 3):
            new_rows = pd.DataFrame( {'id':[i], 'Tamanho':[tam], 'col1':[DCcol[0]], 'op1':[DCop[0]], 'col2':[DCcol[1]], 'op2':[DCop[1]], 'col3':[DCcol[2]], 'op3':[DCop[2]], 'col4':[0], 'op4':[0]} )
            DCs = pd.concat([DCs, new_rows])
        elif(tam == 4):
            new_rows = pd.DataFrame( {'id':[i], 'Tamanho':[tam], 'col1':[DCcol[0]], 'op1':[DCop[0]], 'col2':[DCcol[1]], 'op2':[DCop[1]], 'col3':[DCcol[2]], 'op3':[DCop[2]], 'col4':[DCcol[3]], 'op4':[DCop[3]]} )
            DCs = pd.concat([DCs, new_rows])
    
    df_aux = pd.DataFrame({'parse': aux_parse, 'denial_constraint': aux_DC} )
    df_aux = df_aux.reset_index(drop=True)
    DCs = DCs.reset_index(drop=True)    
    DCs = pd.concat([DCs,df_aux], axis=1)
    print('Dataframe is done!')
    return (DCs)
```

## Coverage functions

```python
def compare(A, B, col, op):
    
    if (op == 'EQUAL'):
        if (A[col] == B[col]): 
            return 1
        return 0            
    elif (op == 'UNEQUAL'):
        if (A[col] != B[col]):
            return 1
        return 0
    elif (op == 'GREATER_EQUAL'):
        if (A[col] >= B[col]):
            return 1
        return 0
    elif (op == 'LESS_EQUAL'):
        if (A[col] <= B[col]):
            return 1
        return 0
    elif (op == 'LESS'):
        if (A[col] < B[col]):
            return 1
        return 0
    elif (op == 'GREATER'):
        if (A[col] > B[col]):
            return 1
        return 0
    
    return -1
```

```python
def Coverage(df, DCs, n_DCs):
    print('Calculating Coverage...')
    cov = []
    doa = []
    for i in range(n_DCs):
        minLen = len(df)
        print ('Processing DC: ', i)
        #Inicialização dos tamanhos
        aux2 = 0
        dOfAprox = 0
        #Comparação entre as tuplas
        K = DCs.iloc[i]                     #K = DC  
        for j in range (len(df)):
            aux1 = 0
            aux3 = 0
            A = df.iloc[j]           #A = Tupla t DB
            for k in range (len(df)):              
                if (j != k):
                    aux = 0             #quantos predicados foram atendidos por aquele par de tupla
                    B = df.iloc[k]       #B = Tupla t' DB
                    #Cálculo Tamanho de validade da operação de cada predicado
                    if (K['Tamanho'] == 1):
                        aux += compare(A, B, K['col1'], K['op1'])
                    elif (K['Tamanho'] == 2):
                        aux += compare(A, B, K['col1'], K['op1'])
                        aux += compare(A, B, K['col2'], K['op2'])
                    elif (K['Tamanho'] == 3):
                        aux += compare(A, B, K['col1'], K['op1'])
                        aux += compare(A, B, K['col2'], K['op2'])
                        aux += compare(A, B, K['col3'], K['op3'])
                    elif (K['Tamanho'] == 4):
                        aux += compare(A, B, K['col1'], K['op1'])
                        aux += compare(A, B, K['col2'], K['op2'])
                        aux += compare(A, B, K['col3'], K['op3'])
                        aux += compare(A, B, K['col4'], K['op4'])
                    else:
                        print('Too much Predicates!')   
                    if (aux < K['Tamanho']):                        #k < Tamanho da DC(Quantidade de predicados)
                        aux1 += 1                                   #absoluto                                                                                       ->|kE|
                        aux3 += (aux+1)/(K['Tamanho'])              #porcentagem respeitada: ( quantidade de predicados respeitados ) / quantidade de predicados    ->|kE|*w(k)    
                    else:
                        dOfAprox += 1                               # somatorio de dOfAprox pra cada linha da esquerda com todas as da direita, exceto qndo (t.direita == t.esquerda)  
            aux2 += (aux3/aux1)                                     # somatorio do calculo da coverage para cada tupla da esquerda por todas as da direita 
        #Calcula coverage de cada predicado
        doa.append(dOfAprox/( len(df) * (len(df)-1) ))
        cov.append(aux2/len(df) )                                   # Média da coverage de todas as linhas    
    
    df_aux = pd.DataFrame( {'coverage':cov, 'deegre of approximation':doa} )    
    df_aux = df_aux.reset_index(drop=True)
    DCs = DCs.reset_index(drop=True)    
    
    DCs = pd.concat([DCs, df_aux], axis=1)
    
    print('Done!')
    return(DCs)
```

## Succinctness functions

```python
def different_elements_list(column, i_begin, i_end):
    unique_list = []
    for j in range(i_begin, i_end):
        if not(column[j] in unique_list):
            unique_list.append( str(column[j]) )
    return len(unique_list)
```

```python
def Succ(DCs, n_DCs):
    print('Calculating Succinctness...')
    succ = []   
    tam_DC = []
    for i in range(n_DCs):        
        print ('Processing DC: ', i)
        K = DCs.iloc[i]                     #K = DC
        #Tamanho total       
        tam = different_elements_list(K['parse'], 0, K['Tamanho']*5)           
        tam_DC.append(tam)
        #Tamanho de todos os predicados
        predicados = []
        for j in range(K['Tamanho']):
            predicados.append(different_elements_list(K['parse'], j*5, (j+1)*5))
        #Calculo da Succinctness de K
        menor_predicado = min(predicados)
        succ.append(menor_predicado/tam_DC[i])
                                       
    df_aux = pd.DataFrame( {'succinctness':succ} )    
    df_aux = df_aux.reset_index(drop=True)
    DCs = DCs.reset_index(drop=True)    
    DCs = pd.concat([DCs, df_aux], axis=1)
    print('Done!')
    return(DCs)
```

## Interestingness functions

```python
def Interestingness (DCs, n_DCs):
    alpha = 0.001                           #Noise Level
    print('Calculating Interestingness...')
    interest = []   
    for i in range(n_DCs):
        print ('Processing DC: ', i)
        K = DCs.iloc[i]                     #K = DC
        interest.append(alpha * K['coverage'] + (1 - alpha) * K['succinctness'])
    DCs['interestingness'] = interest
    print('Done!')
    return(DCs)
```

```python
def top (DCs, label, k):
    if (label != 'deegre of approximation'):
        return( DCs.sort_values(label, ascending=False).head(k) )
    else:
        return( DCs.sort_values(label, ascending=True).head(k) )
```
